***Visualizing Loans Awarded by Kiva***

In this project you'll visualize insights using a dataset from Kaggle. The dataset contains information about loans awarded by the non-profit Kiva.

Using Seaborn, you'll explore the average loan amount by country using aggregated bar charts. You'll also visualize the distribution of loan amounts by project type and gender using box plots and violin plots

# step_two::Import Necessary Python Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Step_three:: Ingest The Data

```python
df=pd.read_csv('C:/Users/amr/Downloads/code_academy _course/seabornProjectCumulative/kiva_data.csv')
df.head(7)
```

# Step_four: Bar Charts

***Create a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.***

```python
import matplotlib.ticker as mtick
f,ax=plt.subplots(figsize=(15,10))
sns.barplot(data=df,x="country",y="loan_amount")
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.show()
```

# Step_five: Learn More By Using hue In Your Visualization

***visualize even more data on one bar plot by visualizing the loan amount by country, and "nesting" by gender***

```python

plt.figure(figsize=(15,10))
sns.set_style("whitegrid")
sns.set_palette("Accent")
plt.title("Average Loan Amounts Given by Kiva, by country, by gender (in $ USD)")
sns.barplot(data=df,x="country",y="loan_amount",hue="gender")
plt.show()
```

# Step_seven: Box Plots With Kiva Data

***make a box plot to compare the distribution of loans by country.***

```python
plt.figure(figsize=(16,10))
sns.boxplot(data=df,x="country",y="loan_amount")
plt.show()
```

# step_nine:: Box Plot by Activity

```python
plt.figure(figsize=(16,10))
sns.boxplot(data=df,x="activity",y="loan_amount")
plt.show()
```

# step_eleven::Violin Plots

***Create a violin plot that visualizes the distribution of loan amount by country***

```python
plt.figure(figsize=(16,10))
sns.violinplot(data=df,x="country",y="loan_amount")
plt.show()
```

# Step_twelve: Split Violin Plots by gender

```python
plt.figure(figsize=(16,10))
sns.set_palette("Spectral")
sns.violinplot(data=df,x="country",y="loan_amount",hue="gender",split=True)
plt.show()
```
